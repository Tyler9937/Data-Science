```
# Conducting further cleaning as to perfect messy merged dataframe
def result_cleaning(x):
  x = str(x).replace("-", " ")
  return x

result['Strain'] = result['Strain'].apply(result_cleaning)
```

```
# Replacing placeholders as None values
result['Medical'] = result['Medical'].replace('Info not available', None, regex=True)
result['Negative'] = result['Negative'].replace('Info not available', None, regex=True)
```
